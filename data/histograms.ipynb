```python
import numpy as np
import matplotlib.pyplot as plt
import ipympl
import imageio.v3 as iio
import skimage
import skimage.draw

%matplotlib widget

#%matplotlib inline
```

```python
plant_seedling = iio.imread(uri="data/plant-seedling.jpg", mode="L")
```

```python
plant_seedling = skimage.util.img_as_float(plant_seedling)

fig, ax = plt.subplots()
plt.imshow(plant_seedling, cmap="gray")
```

```python
# create the histogram
histogram, bin_edges = np.histogram(plant_seedling, bins=256, range=(0,1))
```

```python
histogram
```

```python
# configure and draw the histogram figure

fig, ax = plt.subplots()

plt.title("Grayscale histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.xlim([0.0, 1.0])

plt.plot(bin_edges[0:-1], histogram)
```

# Exercise until 15:50!

```python
plant_mask = np.zeros(shape=plant_seedling.shape, dtype="bool")
rows, columns = skimage.draw.rectangle(start=(199, 410), end=(384, 485))

plant_mask[rows, columns] = True

# DIsplay the mask
fig, ax = plt.subplots()
plt.imshow(plant_mask, cmap="gray")
```

```python
histogram, bin_edges = np.histogram(plant_seedling[plant_mask], bins=256, range=(0.0, 1.0))

plt.figure()
plt.title("Grayscale histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.xlim([0.0, 1.0])
plt.plot(bin_edges[0:-1], histogram)
```

## Color histograms

```python
# Read original picture in full color
plant_seedling_color = iio.imread(uri="data/plant-seedling.jpg")

# Display
fig, ax = plt.subplots()
plt.imshow(plant_seedling_color)
```

```python
plant_seedling_color[:, :, 0]

fig, ax = plt.subplots()
plt.imshow(plant_seedling_color[:, :, 1], cmap='gray')
```

```python
colors = ("red", "green", "blue")

# Create the histogram plot, with three lines, one for each color

fig, ax = plt.subplots()
plt.xlim([0, 256])

for channel_id, color in enumerate(colors):
    histogram, bin_edges = np.histogram(
        plant_seedling_color[:, :, channel_id], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=color)
    
plt.title("Color histogram")
plt.xlabel("Color value")
plt.ylabel("pixel count")
```
